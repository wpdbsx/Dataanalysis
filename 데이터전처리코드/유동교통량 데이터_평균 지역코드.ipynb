```python
#*********GPS
# 필요한 모듈 가져오기 => json 파싱을 위해 들고옴
# 최소한 사용하기 직전에만 모듈 가져오기를 수행하면 된다 / 코드 어느 위치든 상관 없음
import urllib.request
import os
import sys
import json
```

```python
# 바뀔일이 없는 정보는 대문자 변수로 설정하여 상수화
target_url = 'https://dapi.kakao.com/v2/local/search/keyword.json'
KAKAO_API = 'KakaoAK 8496a029241807d997d53db97f23e968'

# 항수화 
# 입력 : 검색어
# 출력 : addr, lat, lng
# 함수명 : getGPSbyKeyword

def getGPSbyKeyword(keyword):
    keyword =  urllib.parse.quote(keyword)
    request = urllib.request.Request(f'{target_url}?query={keyword}')
    
    request.add_header('Authorization', KAKAO_API)
    response = urllib.request.urlopen( request )
    
    if response.getcode() == 200:
        tmp = json.load(response)
        if len(tmp['documents'])==0:
#             print(tmp)
            return ' ',' ',' '
        else:
            addr, lat, lng = tmp['documents'][0]['address_name'], tmp['documents'][0]['y'], tmp['documents'][0]['x']
        
        return addr, lat, lng
    else:
      # 통신 실패 
        addr, lat, lng = ('',0,0)
        return addr, lat, lng
```
